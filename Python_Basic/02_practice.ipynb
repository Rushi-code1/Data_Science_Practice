## Filtering Practise task 


1 Filter by Salary: Create a DataFrame with employee data that includes columns for Name, Salary, and Department. Filter the DataFrame to show all employees with a salary greater than $60,000.

2 Filter by Department: Use the same DataFrame from the first task. Filter the DataFrame to show all employees who work in the 'Sales' department.

3 Filter by Name and Salary: Create a DataFrame with columns Name, Age, and Salary. Filter the DataFrame to find employees whose name starts with the letter 'J' and who have a salary less than $50,000.

4 Filter by Multiple Conditions: Extend the DataFrame from the previous tasks to include a Location column. Filter the DataFrame to show all employees with a salary greater than $55,000 who are located in 'New York'.

```python
import pandas as pd
import numpy as np
# Sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'James', 'Laura', 'Michael', 'Nina', 'Oscar'],
    'Salary': [70000, 55000, 60000, 80000, 45000, 65000, 72000, 48000, 53000, 75000],
    'Department': ['HR', 'Sales', 'Marketing', 'IT', 'Finance', 'Sales', 'HR', 'Finance', 'Marketing', 'IT'],
    'Location': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix', 'New York', 'Chicago', 'Houston', 'New York',
                  'Phoenix'],
    'Age': np.random.randint(20, 60, size=10)
}

df = pd.DataFrame(data)
```

```python
import pandas as pd
import numpy as np

# List of 50 sample names
names = [
    'Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace', 'Hannah', 'Ian', 'Jane',
    'Kyle', 'Laura', 'Mike', 'Nina', 'Oscar', 'Paul', 'Quinn', 'Rachel', 'Steve', 'Tina',
    'Uma', 'Victor', 'Wendy', 'Xander', 'Yara', 'Zane', 'Aaron', 'Bella', 'Caleb', 'Diana',
    'Ethan', 'Fiona', 'George', 'Hazel', 'Isaac', 'Jasmine', 'Kevin', 'Lily', 'Mason', 'Nora',
    'Oliver', 'Piper', 'Quincy', 'Riley', 'Sophia', 'Tyler', 'Ursula', 'Vince', 'Willow', 'Zoe'
]

# Generate random ages and salaries
np.random.seed(0)  # For reproducibility
ages = np.random.randint(20, 60, size=50)  # Random ages between 20 and 60
salaries = np.random.randint(30000, 100000, size=50)  # Random salaries between 30,000 and 100,000

# Create DataFrame
df2 = pd.DataFrame({
    'Name': names,
    'Age': ages,
    'Salary': salaries
})

print("DataFrame:")
print(df2)
```

```python
df2[(df2["Name"].str.startswith("J")) & (df2["Salary"]<50000)]
```

```python
df[(df["Location"]=="New York") & (df["Salary"] > 50000)]
```

```python
# Filter for Employees in HR or IT Department:

# Get all rows where the Department is either 'HR' or 'IT'.


df[(df["Department"] == 'HR') | (df["Department"] == "IT")]
```

```python
# Filter for Employees with Salary Not Equal to 75,000:

# Get all rows where the Salary is not equal to 75,000.

df[df["Salary"] != 75000]
```

```python

# Filter for Age Less Than 30 or Salary Greater Than 90,000:

# Get all rows where the Age is less than 30 or the Salary is greater than 90,000.

df2[(df2["Salary"] > 90000 ) | (df2["Age"] < 30)]
```

```python
# Filter for Employees Whose Names Have Exactly 5 Characters:

# Get all rows where the Name has exactly 5 characters.

df[df["Name"].str.len() == 5]
```

```python
# Filter for Employees Not Located in New York:

# Get all rows where the Location is not 'New York'.

df[df["Location"] != 'New York']
```

```python
# Filter for Employees with Age Between 25 and 35 and Working in Finance:

# Get all rows where the Age is between 25 and 35 (inclusive) and the Department is 'Finance'.
df[((df["Age"] >= 25) & (df["Age"] <= 35)) & (df['Department'] == 'Finance')]
```

```python
# Filter for Employees with Names Containing an 'e' and Located in Chicago:

# Get all rows where the Name contains the letter 'e' and the Location is 'Chicago'.

df[(df["Name"].str.contains("e")) & (df["Location"]=="Chicago")]
```
