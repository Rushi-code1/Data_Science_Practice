## Data Visualization

```python

# Load your data
data =df

# Set plot style for better aesthetics
sns.set(style="whitegrid")

# Create subplots to visualize the distributions and outliers for different columns
fig, axs = plt.subplots(3, 2, figsize=(15, 15))

# Plot Histogram and Box Plot for Volume
sns.histplot(data['Volume'], bins=10, kde=True, ax=axs[0, 0],color="orange")
axs[0, 0].set_title('Distribution of Volume')
sns.boxplot(x=data['Volume'], ax=axs[0, 1])
axs[0, 1].set_title('Box Plot of Volume')

# Plot Histogram and Box Plot for Avg Price
sns.histplot(data['Avg Price'], bins=10, kde=True, ax=axs[1, 0],color="orange")
axs[1, 0].set_title('Distribution of Avg Price')
sns.boxplot(x=data['Avg Price'], ax=axs[1, 1])
axs[1, 1].set_title('Box Plot of Avg Price')

# Plot Histogram and Box Plot for Net Sales Value
sns.histplot(data['Net Sales Value'], bins=10, kde=True, ax=axs[2, 0],color="orange")
axs[2, 0].set_title('Distribution of Net Sales Value')
sns.boxplot(x=data['Net Sales Value'], ax=axs[2, 1])
axs[2, 1].set_title('Box Plot of Net Sales Value')

plt.tight_layout()
plt.show()
```

```python
# Calculate the total sales value per city
sales_per_city = data.groupby('Brand')['Total Sales Value'].sum().reset_index()

# Sort by total sales value for better visualization
sales_per_city = sales_per_city.sort_values('Total Sales Value', ascending=False)

# Create the bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x='Brand', y='Total Sales Value', data=sales_per_city)
plt.title('Total Sales Value by Brand')
plt.xlabel('Brand')
plt.ylabel('Total Sales Value')
plt.xticks(rotation=40)  # Rotate city names for better readability
plt.show()
```

## Conversion of Categorical Data into Dummy Variables

1. Need for Converting Categorical Data into Dummy Variables
Why Convert Categorical Data?

Machine Learning Models: Most ML algorithms (like linear regression, logistic regression, decision trees, etc.) require numerical input. Categorical variables are nominal (having no intrinsic order) or ordinal (having a clear ordering) and must be converted to numbers.

Avoiding Misinterpretation: If you assign arbitrary numerical values to categories (e.g., 1, 2, 3), models may interpret these numbers as having an order or relationship that doesn't exist, leading to poor performance.

Handling Multiple Categories: One-hot encoding creates a new binary column for each unique category level, allowing the model to treat each category independently.

2. Applying One-Hot Encoding
One-hot encoding creates a new binary column for each category level in a categorical variable. For example, if you have a column Color with values Red, Blue, and Green, one-hot encoding would create three new columns: Color_Red, Color_Blue, and Color_Green.

```python
# Identify categorical columns
categorical_columns = data.select_dtypes(include=['object', 'category']).columns

# Apply one-hot encoding
data_encoded = pd.get_dummies(data, columns=categorical_columns, drop_first=False)

# Display a portion of the transformed dataset
print(data_encoded.head(6))
```
