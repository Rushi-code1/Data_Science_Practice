# 3. Feature Engineering:


```python
df['work_life_ratio'] = df['hours_per_week'] / df['age']
df['age_education_interaction'] = df['age'] * df['education_num']

bins = [0, 25, 45, 65, 100]
labels = ['young', 'middle_aged', 'senior', 'elderly']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels)
```

```python
df
```

## Rationale:

#### * Work-Life Ratio: Divide hours-per-week by age to capture how much time each person is spending working relative to their age.
#### * The age-education interaction feature helps the model understand the joint influence of education and experience on salary, giving it a better chance to model the true salary distribution in the dataset.
#### * Age group binning helps transform a continuous variable (age) into a more meaningful categorical variable that can enhance the predictive power of your model by better capturing the relationship between age and income.

# Scaling
*	Apply scaling techniques to numerical features:

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()  # or MinMaxScaler()
scaled_data = scaler.fit_transform(df[['age', 'hours_per_week']])  # Example of scaling "age" and "hours-per-week"
```

•	Discuss the scenarios where each scaling technique is preferred and why.
StandardScaler is preferred in the following scenarios:

* When the data follows a Gaussian distribution, as it scales the data to have a mean of 0 and a standard deviation of 1.
* When the algorithm used is sensitive to the scale of the data, such as in neural networks or support vector machines.
* When the data has outliers, as it is robust to outliers and can handle them effectively.

MinMaxScaler is preferred in the following scenarios:

* When the data does not follow a Gaussian distribution, as it scales the data to a specific range (usually between 0 and 1) without making any assumptions about the distribution.
* When the algorithm used is not sensitive to the scale of the data, such as in decision trees or random forests.
* When the data needs to be scaled to a specific range for interpretation or visualization purposes.

In general, StandardScaler is a good default choice when the data is normally distributed and the algorithm is sensitive to scale. MinMaxScaler is a good choice when the data is not normally distributed or when the algorithm is not sensitive to scale.

## 2. Encoding Techniques

*	Apply One-Hot Encoding to categorical variables with less than 5 categories.

```python
for i in df.drop(columns='income').select_dtypes(include=['object']):
        if len(df[i].unique())<5 :
                df = pd.get_dummies(df, columns=[i])
```

*	Use Label Encoding for categorical variables with more than 5 categories.

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for i in df.select_dtypes(include=['object', "category"]):
      df[i] = le.fit_transform(df[i])
```
